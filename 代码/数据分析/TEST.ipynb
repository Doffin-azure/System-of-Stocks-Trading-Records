```python
import pandas as pd

# 英文列名列表
english_names = [
    "tradedate", "OrigTime", "SendTime", "recvtime", "dbtime", "ChannelNo", "MDStreamID", 
    "ApplSeqNum", "SecurityID", "SecurityIDSource", "Price", "OrderQty", "TransactTime", 
    "Side", "OrderType", "ConfirmID", "Contactor", "ContactInfo", "ExpirationDays", "ExpirationType"
]

# 读取文件，批次大小为1000000
batch_size = 1000000
lines_list = []

def process_batch(batch_df):
    # 只选择需要的列
    easy_order_table = batch_df[["ChannelNo", "ApplSeqNum", "TransactTime"]]
    
    # 将列转换为 Int64 类型
    easy_order_table = easy_order_table.astype({"ChannelNo": "Int64", "ApplSeqNum": "Int64"})
    
    # 按照 'ChannelNo' 和 'ApplSeqNum' 排序
    easy_order_table = easy_order_table.sort_values(by=["ChannelNo", "ApplSeqNum"])

    # 按照 'ChannelNo' 划分
    ChannelNo = easy_order_table["ChannelNo"].unique()

    # 检查每个 'ChannelNo' 的 'ApplSeqNum' 是否升序
    for i in ChannelNo:
        is_ascending = easy_order_table[easy_order_table["ChannelNo"] == i]["ApplSeqNum"].is_monotonic_increasing
        if not is_ascending:
            print(f"ChannelNo {i} does not have an increasing ApplSeqNum.")
    
    # 返回处理后的数据框，供下一次批次处理使用
    return easy_order_table

# 逐行读取文件并按批次处理
with open('20190102/pm_hq_order_spot.txt', 'r') as f:
    count = 0
    while True:
        count += 1
        
        # 读取每一批数据
        lines_batch = []
        for _ in range(batch_size):
            line = f.readline()
            if not line:  # 文件结束时退出
                break
            lines_batch.append(line.split('\t'))
        
        if lines_batch:
            # 将当前批次转换为 DataFrame
            batch_df = pd.DataFrame(lines_batch, columns=english_names)
            
            # 处理当前批次
            process_batch(batch_df)
            print(f"Processed {count} lines")
        
        if not lines_batch:  # 文件读取完毕
            print("It's over")
            print(f"Total lines processed: {count}")
            break
```
